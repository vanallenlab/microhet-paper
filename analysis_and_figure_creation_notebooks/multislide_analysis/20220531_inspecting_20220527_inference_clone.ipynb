```python
df_pivot = df.pivot_table(index=['slide_id','x','y','full_path','dataset'], columns='model_id', values='class_1_model_prob')
df_pivot = df_pivot.reset_index().set_index(['slide_id','x','y'])
g_cols = list(filter(lambda x: 'profile' in x, df_pivot.columns))
df_pivot['ensemble_prob_g4_not_g2'] = df_pivot[g_cols].mean(1)
```

```python
df_pivot = df_pivot.reset_index().rename(columns={'tvnt':'prob_tumor', 'ensemble_prob_g4_not_g2':'prob_g4_not_g2','slide_id':'unique_id',})
```

```python
df_pivot
```

```python
inference_raw = {}
inference_raw['denize_multi'] = df_pivot
```

```python
from collections import defaultdict

META_VARS = ['stroma', 'pred_g2', 'intermediate_grade', 'pred_g4']
N_NEIGHBORS = 4
TUMOR_CUTOFF = 0.5
LOWER_CUTOFF = 1/3.
UPPER_CUTOFF = 2/3.
print(f'using p(Tumor) cutoff of {TUMOR_CUTOFF}')

###########
all_results = defaultdict(list)
print('using aggregate mean pred. over all 4 models from 20210312 checkpoints ')
for key,df in inference_raw.items():
    print(key)
    for slide_id, slide_df in df.groupby('unique_id'):
        try:
            temp = run_inference_smoothing(slide_df, dist_type='uniform', n_neighbors=N_NEIGHBORS, tumor_cutoff=TUMOR_CUTOFF, lower_grade_cutoff=LOWER_CUTOFF, upper_grade_cutoff=UPPER_CUTOFF)

            all_results[key].append(temp)

        except Exception as e:
            print(e)
            print(f'Error with {slide_id}')
            
combined = []
for key,val in all_results.items():
    print(pd.concat(val).shape)
    temp = pd.concat(val)
    temp['cohort'] = key
    combined.append(temp)
combined = pd.concat(combined)
combined_results = combined[['unique_id','x','y','prob_tumor','prob_g4_not_g2','smoothed_prob_tumor','smoothed_prob_g4_not_g2','meta','cohort']]
```
